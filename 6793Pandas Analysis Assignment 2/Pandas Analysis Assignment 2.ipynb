```python
df_jobrole = data[['JobRole','DistanceFromHome','Attrition']]
df_jobrole.head()
```

```python
plt.figure(figsize=(20,10))
sns.barplot(x='JobRole', y='DistanceFromHome', hue='Attrition', data=df_jobrole)
plt.title("Rate of attrition by distance and job role")
plt.xlabel("Job Role")
plt.ylabel("Distance from Home")
```

```python
attrition_yes = df_jobrole[['DistanceFromHome', 'JobRole', 'Attrition']][df_jobrole['Attrition'] == 'Yes']
attrition_no = df_jobrole[['DistanceFromHome', 'JobRole', 'Attrition']][df_jobrole['Attrition'] == 'No']
attrition_yes
```

```python
#keys = role for df_jobrole['JobRole'].unique(), df_jobrole.groupby(['Attrition'])
attyes = attrition_yes.groupby(['DistanceFromHome', 'JobRole']).count().reset_index()
attyes
```

```python
distance_yes = attrition_yes[['DistanceFromHome','Attrition']]
distypes = distance_yes.groupby(['DistanceFromHome']).count().reset_index()
distypes

```

```python
plt.figure(figsize=(20,10))
sns.barplot(x='DistanceFromHome', y='Attrition', data=distypes)
plt.title("Rate of attrition by distance")
plt.xlabel("Distance")
plt.ylabel("No.of employees who stayed")
```

```python
role_yes = attrition_yes[['JobRole','Attrition']]
roletypes = role_yes.groupby(['JobRole']).count().reset_index()
roletypes
```

```python
plt.figure(figsize=(20,10))
sns.barplot(x='JobRole', y='Attrition', data=roletypes)
plt.title("Rate of attrition by Job Role")
plt.xlabel("Job Role")
plt.ylabel("No.of employees who stayed")
```
